### Η εξίσωση του Ραντάρ

Θεωρούμε ένα ραντάρ με μια ομοιοκατευθυντική κεραία (αυτή που εκπέμπει ενέργεια εξίσου προς όλες τις κατευθύνσεις). Δεδομένου ότι αυτού του είδους οι κεραίες έχουν σφαιρικό διάγραμμα ακτινοβολίας, μπορούμε να ορίσουμε την μέγιστη πυκνότητα ισχύος (ισχύς ανά μονάδα επιφάνειας) σε οποιοδήποτε σημείο στο χώρο ως

$$
PD = \frac{Peak \ transmitted \ power}{area \ of \ a \ sphere}  \qquad watts/m^2
$$

Η πυκνότητα ισχύος σε απόσταση R από το ραντάρ (υποθέτοντας διάδοση σε μέσο χωρίς απώλειες) είναι

$$
PD = \frac{P_t}{4\pi R^2}
$$

όπου $P_t$ είναι η μέγιστη εκπεμπόμενη ισχύς και $4 \pi R^2 $ είναι η επιφάνεια της σφαίρας ακτίνας R. Τα ραντάρ χρησιμοποιούν κατευθυντικές κεραίες, προκειμένου να αυξήσουν την πυκνότητα ισχύος προς μια συγκεκριμένη κατεύθυνση. Οι κατευθυντικές κεραίες συνήθως χαρακτηρίζονται από το κέρδος κεραίας G και την ενεργό επιφάνεια της κεραίας $A_e$. Σχετίζονται με

$$
G = \frac{4\pi A_e}{λ^2}
$$

όπου λ είναι το μήκος κύματος. Η σχέση ανάμεσα στην ενεργό επιφάνεια της κεραίας $A_e$ και στην πραγματική της επιφάνεια A είναι 
$$
A_e= ρΑ, \qquad 0 \leq ρ \leq 1
$$

όπου ο συντελεστής ρ αναφέρεται ως απόδοση κεραίας και οι καλές κεραίες απαιτούν να προσεγγίζει τη μονάδα. Εδώ θα υποθέσουμε, ότι είναι μονάδα και άρα η ενεργός με την πραγματική επιφάνεια είναι το ίδιο.
Θα υποθέσουμε επίσης ότι οι κεραίες έχουν το ίδιο κέρδος στην εκπομπή και στη λήψη. Στην πράξη, κάτι τέτοιο είναι ευρέως αποδεκτό, εφόσον χρησιμοποιείται η ίδια κεραία.
Το κέρδος της κεραίας σχετίζεται επίσης με το οριζόντιο (αζιμούθιο) και κατακόρυφο εύρος της δέσμης της κεραίας κατά


$$
G = k \frac{4\pi}{θ_e θ_a}
$$

όπου $k \leq 1$ και εξαρτάται από το φυσικό σχήμα της κεραίας, ενώ $ θ_e, \quad θ_a είναι οι γωνίες κατακόρυφου και οριζόντιου εύρους της δέσμης αντίστοιχα σε ακτίνια. Μια εξαιρετική προσέγγιση της ανωτέρω σχέσης, που εισήγαγε ο Stutzman και αναφέρεται στο βιβλίο του Skolnik είναι

$$
G = \frac{2600}{θ_e θ_a}
$$

όπου στην περίπτωση αυτή το κατακόρυφο και το οριζόντιο εύρος δέσμης δίνονται σε μοίρες.

Η πυκνότητα ισχύος σε απόσταση R από ένα ραντάρ με κεραία κέρδους G δίνεται από

$$
PD = \frac{P_t G}{4\pi R^2}
$$

Όταν η ενέργεια που εκπέμπεται από το ραντάρ προσκρούει σε έναν στόχο, τα επαγόμενα επιφανειακά ρεύματα σε αυτόν τον στόχο εκπέμπουν ηλεκτρομαγνητική ενέργεια προς όλες τις κατευθύνσεις. Το ποσό της ακτινοβολούμενης ενέργειας είναι ανάλογo με το μέγεθος του στόχου, τον προσανατολισμό του, το σχήμα του και το υλικό, παράμετροι που συγκεντρώνονται όλες μαζί σε μία παράμετρο χαρακτηριστική για το στόχο, που λέγεται Ενεργός Διατομή Ραντάρ - Radar Cross Section (RCS) και συμβολίζεται με σ.
Το RCS ορίζεται ως ο λόγος της ισχύος που ανακλάται πίσω στο ραντάρ ως προς την πυκνότητα ισχύος που προσπίπτει στο στόχο,

$$
σ = \frac {P_r}{P_D} \quad m^2
$$

όπου $P_r$ είναι η ισχύς που ανακλάται από το στόχο. Έτσι, η ολική ισχύς που φτάνει πίσω στο δέκτη του ραντάρ από την κεραία είναι 

$$
P_{DR} = \frac{P_t G σ}{(4\pi R^2)^2}A_e
$$

Χρησιμοποιώντας της σχέση για την ενεργό επιφάνεια $A_e$ της κεραίας, προκύπτει ότι

$$
P_{DR} = \frac{P_t G^2 λ^2 σ}{(4\pi)^3 R^4}
$$

Αν με $S_{min}$ συμβολίσουμε το ελάχιστο εντοπιζόμενο σήμα στο δέκτη, αυτό αντιστοιχεί προφανώς στη μέγιστη απόσταση εντοπισμού $R_{max}$, που δίνεται από τη σχέση

$$
R_{max} = \left(\frac{P_t G^2 λ^2 σ}{(4\pi)^3 S_{min}}\right)^\frac{1}{4}
$$

Στην πράξη τα επιστρεφόμενα σήματα που λαμβάνονται από το ραντάρ περιλαμβάνουν θόρυβο, που εισάγει ανεπιθύμητες τάσεις σε όλες τις συχνότητες ραντάρ. Ο θόρυβος είναι τυχαίος στη φύση του και μπορεί να περιγραφεί από τη Φασματική Πυκνότητα Ισχύος του (PSD). Η ισχύς θορύβου Ν είναι συνάρτηση του εύρους ζώνης λειτουργίας του ραντάρ Β. Ακριβέστερα

$$
N = Noise \ PSD \ Χ \ B
$$

Η εισερχόμενη ισχύς θορύβου σε μια κεραία χωρίς απώλειες

$$
N_i = kT_eB
$$

όπου $ k = 1.38 x 10^-23$ Joule/deg Kelvin είναι η σταθερά του Boltzman και $T_e$ είναι η ισοδύναμη θερμοκρασία θορύβου σε βαθμούς Kelvin. Είναι πάντα επιθυμητό το ελάχιστο εντοπιζόμενο σήμα $S_{min}$ να είναι ισχυρότερο της ισχύος του θορύβου. Η απόδοση του δέκτη καθόσον ως ηλεκτρονική συσκευή προσθέτει περισσότερο θόρυβο και επιβαρύνει την εντοπιστική ικανότητα συνήθως περιγράφεται με ένα δείκτη θορύβου Noise Figure (NF) που ορίζεται ως

$$
F = \frac{SNR_i}{SNR_o} = \frac{S_i/N_i}{S_o/N_o}
$$

όπου $SNR_i, SNR_o$ είναι ο σηματοθορυβικός λόγος στην είσοδο και στην έξοδο του δέκτη αντίστοιχα. Χρησιμοποιώντας τις σχέσεις αυτές προκύπτει ότι
$$
S_i = kT_eBF(SNR_o)
$$
και άρα το ελάχιστο εντοπιζόμενο σήμα στην είσοδο του δέκτη εκφρασμένο συναρτήσει του ελάχιστου σηματοθορυβικού λόγου στην έξοδο του δέκτη δίνεται από
$$
S_{min} = kT_eBF(SNR_o)_{min}
$$

Το κατώφλι εντοπισμού του ραντάρ είναι ακριβώς ο ελάχιστος σηματοθορυβικός λόγος στη έξοδο του δέκτη $(SNR_o)_{min}$. Η μέγιστη απόσταση εντοπισμού του ραντάρ δίνεται από

$$
R_{max} = \left[ \frac{P_t\,G^2\,\lambda^2\,\sigma}{(4\pi)^3\,k\,T_e\,B\,F\,({SNR}_o)_{min}} \right]^{1/4} \hspace{0.5in} \text{(m)}
$$

η ισοδύναμα ο ελάχιστος σηματοθορυβικός λόγος
$$
({SNR}_o)_{min} = \frac{P_t\,G^2\,\lambda^2\,\sigma}{(4\pi)^3\,k\,T_e\,B\,F\,R_{max}^4} \hspace{0.5in} \text{(m)}
$$

Οι απώλειες του ραντάρ εκφράζονται συνήθως ως ένας συντελεστής L που μειώνει περεταίρω το λαμβανόμενο σήμα και κατ' επέκταση τον ελάχιστο σηματοθορυβικό λόγο. Η σχέση τότε δίνεται από:
$$
({SNR}_o)_{min} = \frac{P_t\,G^2\,\lambda^2\,\sigma}{(4\pi)^3\,k\,T_e\,B\,F\,L\,R_{max}^4} \hspace{0.5in} \text{(m)}
$$

Κατά τη λειτουργία εντοπισμού του ραντάρ (ανίχνευση) συνήθως προδιαγράφεται για ένα ραντάρ ότι ένας ορισμένος στόχος (με μια συγκεκριμένη τιμή RCS) εντοπίζεται βάσει μιας προκαθορισμένης πιθανότητας εντοπισμού σε μια δεδομένη απόσταση. Η πιθανότητα εντοπισμού χρησιμοποιείται για να υπολογιστεί το απαιτούμενο SNR εντοπισμού. Μόλις υπολογιστεί το απαιτούμενο SNR, η προηγούμενη εξίσωση μπορεί να χρησιμοποιηθεί για την εύρεση του καταλληλότερου παλμού (ή κυματομορφής) που επιτυγχάνει το απαιτούμενο SNR. Μπορεί να συμβαίνει ότι καμία από τις διαθέσιμες κυματομορφές ραντάρ δεν μπορεί να εγγυηθεί το ελάχιστο απαιτούμενο SNR για τη συγκεκριμένη τιμή RCS στη δεδομένη απόσταση. Σε αυτή την περίπτωση, το ραντάρ πρέπει να περιμένει έως ότου ο στόχος πλησιάσει αρκετά στην εμβέλεια για να γίνει ανίχνευση, διαφορετικά μπορεί να χρησιμοποιηθεί η ολοκλήρωση παλμών (συνμφασική ή μη συμφασική). \
Γενικά ισχύει σχετικά με το μήκος παλμού και το εύρος ζώνης του ραντάρ η σχέση $B= \frac{1}{τ}$ Η ισχύς εκπμπής $P_t$ σχετίζεται με τη μέση ισχύ ενός ραντάρ με τη σχέση $P_t = P_{av}T/t$, όπου Τ είναι το διάστημα επανάληψης των παλμών (PRI).
